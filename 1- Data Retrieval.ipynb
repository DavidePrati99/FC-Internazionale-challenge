# Scraping da SoFIFA

Effettuo invece uno scraping da SoFIFA per estrarre dati quali altezza, peso, piede preferito e valore economico del giocatore. Il codice potrebbe non funzionare per motivi legati allo scraping.

```python
base_url = "https://sofifa.com/players?type=all&lg%5B0%5D=13&lg%5B1%5D=16&lg%5B2%5D=19&lg%5B3%5D=31&lg%5B4%5D=53&pn%5B0%5D=5&showCol%5B0%5D=ae&showCol%5B1%5D=hi&showCol%5B2%5D=wi&showCol%5B3%5D=pf&showCol%5B4%5D=vl"

# Funzione per estrarre dati da una pagina specifica
def scrape_page(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }

    with requests.Session() as session:
        response = session.get(url, headers=headers)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            player_data = []

            main_element = soup.find('main')
            if main_element:
                article_element = main_element.find('article')
                if article_element:
                    table_element = article_element.find('table')
                    if table_element:
                        tbody_element = table_element.find('tbody')
                        if tbody_element:
                            for row in tbody_element.find_all('tr'):
                                columns = row.find_all('td')

                                # Estraggo le colonne di interesse
                                player_name = columns[1].text.strip()
                                age = columns[2].text.strip()
                                club = columns[3].text.strip()
                                height = columns[4].text.strip()
                                weight = columns[5].text.strip()
                                foot = columns[6].text.strip()
                                value = columns[7].text.strip()

                                player_data.append({
                                    'Player': player_name,
                                    'Age': age,
                                    'Squad': club,
                                    'Height': height,
                                    'Weight': weight,
                                    'Foot': foot,
                                    'Value': value
                                })

            return player_data
        else:
            print(f"Errore nella richiesta: {response.status_code}")
            return None

# Funzione per scrapare tutte le pagine
def scrape_all_pages(base_url, num_pages):
    all_player_data = []
    for offset in range(0, num_pages * 60, 60):
        url = f"{base_url}&offset={offset}"
        player_data = scrape_page(url)
        if player_data:
            all_player_data.extend(player_data)

    return all_player_data

# Scrap0 tutte le pagine
num_pages_to_scrape = 11  # Specifico il numero massimo di pagine da estrarre
all_data = scrape_all_pages(base_url, num_pages_to_scrape)

# Creo un DataFrame pandas con i dati estratti
df = pd.DataFrame(all_data)

df
```
