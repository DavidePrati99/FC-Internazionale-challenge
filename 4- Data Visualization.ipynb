```python
file_path = '/content/gdrive/MyDrive/FC Internazionale Data Science Challenge/df_per_viz.csv'
df = pd.read_csv(file_path, sep=',')
df['Foot'] = df['Foot'].replace({1: 'Left', 0: 'Right'})
df
```

# Prima coppia di scatterplot

La prima coppia di scatterplot riguarda le caratteristiche fisiche che costituiscono il primo score. Il primo scatterplot è definito su tutto il dataset, il secondo solo sui 5 giocatori selezionati



```python
# Altezza minima massima e mediana
altezza_minima = df['Height'].min()
altezza_massima = df['Height'].max()
altezza_mediana = df['Height'].median()

# Peso minimo massimo e mediano
peso_minimo = df['Weight'].min()
peso_massimo = df['Weight'].max()
peso_mediano = df['Weight'].median()

print("Altezza minima:", altezza_minima)
print("Altezza massima:", altezza_massima)
print("Altezza mediana:", altezza_mediana)
print("Peso minimo:", peso_minimo)
print("Peso massimo:", peso_massimo)
print("Peso mediano:", peso_mediano)
```

```python
# Estraggo i giocatori dell'Inter, per vedere quanti sono
inter_players = df[df['Actual Squad'] == 'Inter']
inter_players
```

```python
# Definisco la palette di colori che utilizzerò
palette = sns.color_palette("Dark2")
```

```python
# Definisco la dimensione dei pallini proporzionale ai valori della colonna 'Value (M)'
sizes = df['Value (M)'] * 20

# Definisco i colori in base al valore di 'Foot' utilizzando la palette
colors = df['Foot'].map({'Left': palette[0], 'Right': palette[1]})

# Imposto il colore nero per 'Francesco Acerbi' e blu per gli altri giocatori dell'Inter
colors[df['Player'] == 'Francesco Acerbi'] = 'black'
colors[df['Actual Squad'] == 'Inter'] = inter_color

# Creo lo scatterplot con dimensioni proporzionali
plt.figure(figsize=(10, 7))
plt.scatter(df['Height'], df['Weight'], c=colors, s=sizes, alpha=0.8)

# Metto in risalto Francesco Acerbi
acerbi_index = df[df['Player'] == 'Francesco Acerbi'].index[0]  # Ottieni l'indice di Francesco Acerbi
plt.scatter(df['Height'][acerbi_index], df['Weight'][acerbi_index], c='black', s=sizes[acerbi_index], alpha=0.8)

# Aggiungo il punto per la posizione di (altezza_mediana, peso_mediano)
plt.scatter(altezza_mediana, peso_mediano, c='yellow', s=20)

# Aggiungo etichette agli assi
plt.xlabel('Altezza')
plt.ylabel('Peso')

# Aggiungo titolo
plt.title('Scatterplot Altezza vs Peso, con dimensione data dal valore di mercato')

# Aggiungo legenda per il piede del giocatore
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[0], markersize=10, label='Mancino'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[1], markersize=10, label='Destro'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='black', markersize=10, label='Francesco Acerbi'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=inter_color, markersize=10, label='Altri giocatori dell\'Inter')])

plt.show()
```

```python
# Definisco la dimensione dei pallini proporzionale ai valori della colonna 'Value (M)'
sizes = df_simili['Value (M)'] * 20

# Definisco i colori in base al valore di 'Foot' utilizzando la palette
colors = df_simili['Foot'].map({'Left': palette[0], 'Right': palette[1]})

# Imposto il colore nero per 'Francesco Acerbi' e blu per gli altri giocatori dell'Inter
colors[df_simili['Player'] == 'Francesco Acerbi'] = 'black'
colors[df_simili['Actual Squad'] == 'Inter'] = inter_color

# Creo lo scatterplot con dimensioni proporzionali
plt.figure(figsize=(10, 7))
plt.scatter(df_simili['Height'], df_simili['Weight'], c=colors, s=sizes, alpha=0.8)

# Metto in risalto Francesco Acerbi
acerbi_index = df_simili[df_simili['Player'] == 'Francesco Acerbi'].index[0]  # Ottieni l'indice di Francesco Acerbi
plt.scatter(df_simili['Height'][acerbi_index], df_simili['Weight'][acerbi_index], c='black', s=sizes[acerbi_index], alpha=0.7)

# Aggiungo il punto per la posizione di (altezza_mediana, peso_mediano)
plt.scatter(altezza_mediana, peso_mediano, c='yellow', s=30)

# Aggiungo etichette agli assi
plt.xlabel('Altezza')
plt.ylabel('Peso')

# Aggiungo il nome del giocatore vicino al pallino
for index, row in df_simili.iterrows():
    if row['Player'] != 'Francesco Acerbi':  # Evita di scrivere il nome se è "Francesco Acerbi"
        player_name = ' '.join(row['Player'].split()[1:])  # Prendi tutte le parole tranne la prima
        plt.text(row['Height'], row['Weight'], player_name, fontsize=12, ha='right', va='top')

# Imposto i limiti degli assi
plt.xlim(altezza_minima-1, altezza_massima+1)
plt.ylim(peso_minimo-2, peso_massimo+2)

# Aggiungo titolo
plt.title('Scatterplot nel caso dei 5 giocatori più simili')

# Aggiungo legenda per il piede del giocatore
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[0], markersize=10, label='Mancino'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[1], markersize=10, label='Destro'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='black', markersize=10, label='Francesco Acerbi'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=inter_color, markersize=10, label='Altri giocatori dell\'Inter')])

plt.show()
```

# Seconda coppia di scatterplot

La seconda coppia di scatterplot riguarda il contrasto tra il primo e il secondo score. Il primo scatterplot è definito su tutto il dataset, il secondo solo sui 5 giocatori selezionati

```python
# Score fisico minimo massimo medio e mediano
score_fisico_minimo = df['score_complessivo'].min()
score_fisico_massimo = df['score_complessivo'].max()
score_fisico_medio = round(df['score_complessivo'].mean(), 4)
score_fisico_mediano = round(df['score_complessivo'].median(), 4)

# Score prestazionale minimo massimo medio e mediano
score_prestazionale_minimo = round(df['similarita_coseno'].min(), 4)
score_prestazionale_massimo = df['similarita_coseno'].max()
score_prestazionale_medio = round(df['similarita_coseno'].mean(), 4)
score_prestazionale_mediano = round(df['similarita_coseno'].median(), 4)

print("Score fisico minimo:", score_fisico_minimo)
print("Score fisico massimo:", score_fisico_massimo)
print("Score fisico medio:", score_fisico_medio)
print("Score fisico mediano:", score_fisico_mediano)
print("Score prestazionale minimo:", score_prestazionale_minimo)
print("Score prestazionale massimo:", score_prestazionale_massimo)
print("Score prestazionale medio:", score_prestazionale_medio)
print("Score prestazionale mediano:", score_prestazionale_mediano)
```

```python
# Definisco la dimensione dei pallini proporzionale ai valori della colonna 'Value (M)'
sizes = df['Value (M)'] * 30

# Definisco i colori in base al valore di 'Foot' utilizzando la palette
colors = df['Foot'].map({'Left': palette[0], 'Right': palette[1]})

# Imposto il colore nero per 'Francesco Acerbi'
colors[df['Player'] == 'Francesco Acerbi'] = 'black'

# Imposto il colore blu per gli altri giocatori dell'Inter che non sono Acerbi
colors[(df['Actual Squad'] == 'Inter') & (df['Player'] != 'Francesco Acerbi')] = 'blue'

# Creo lo scatterplot
plt.figure(figsize=(10, 7))
plt.scatter(df['score_complessivo'], df['similarita_coseno'], c=colors, s=sizes, alpha=0.8)

# Aggiungo il punto per la posizione di (score_fisico_mediano, score_prestazionale_mediano) e (score_fisico_medio, score_prestazionale_medio)
plt.scatter(score_fisico_mediano, score_prestazionale_mediano, c='yellow', s=20)
plt.scatter(score_fisico_medio, score_prestazionale_medio, c='red', s=20)

# Aggiungo etichette agli assi
plt.xlabel('Score fisico')
plt.ylabel('Score prestazionale')

# Aggiungo titolo
plt.title('Scatterplot Score fisico vs prestazionale')

# Aggiungo legenda per il piede del giocatore
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[0], markersize=10, label='Mancino'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[1], markersize=10, label='Destro'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='black', markersize=10, label='Francesco Acerbi'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=inter_color, markersize=10, label='Altri giocatori dell\'Inter')])

plt.show()
```

```python
# Definisco la dimensione dei pallini proporzionale ai valori della colonna 'Value (M)'
sizes = df_simili['Value (M)'] * 30

# Definisco i colori in base al valore di 'Foot' utilizzando la palette
colors = df_simili['Foot'].map({'Left': palette[0], 'Right': palette[1]})

# Imposto il colore nero per 'Francesco Acerbi'
colors[df_simili['Player'] == 'Francesco Acerbi'] = 'black'

# Imposto il colore blu per gli altri giocatori dell'Inter che non sono Acerbi
colors[(df_simili['Actual Squad'] == 'Inter') & (df_simili['Player'] != 'Francesco Acerbi')] = 'blue'

# Creo lo scatterplot
plt.figure(figsize=(10, 7))
plt.scatter(df_simili['score_complessivo'], df_simili['similarita_coseno'], c=colors, s=sizes, alpha=0.8)

# Aggiungo il punto per la posizione di (score_fisico_mediano, score_prestazionale_mediano) e (score_fisico_medio, score_prestazionale_medio)
plt.scatter(score_fisico_mediano, score_prestazionale_mediano, c='yellow', s=20)
plt.scatter(score_fisico_medio, score_prestazionale_medio, c='red', s=20)

# Aggiungo etichette agli assi
plt.xlabel('Score fisico')
plt.ylabel('Score prestazionale')

# Aggiungo il nome del giocatore vicino al pallino
for index, row in df_simili.iterrows():
    if row['Player'] != 'Francesco Acerbi':  # Evito di scrivere il nome se è "Francesco Acerbi"
        player_name = ' '.join(row['Player'].split()[1:])  # Prendo tutte le parole tranne la prima
        if index % 2 == 0:  # Per evitare si sovrappongano le scritte
            ha = 'right'
            va = 'bottom'
        else:
            ha = 'left'
            va = 'top'
        plt.annotate(player_name, (row['score_complessivo'], row['similarita_coseno']), fontsize=12, ha=ha, va=va)

# Imposto i limiti degli assi
plt.xlim(score_fisico_minimo-0.05, score_fisico_massimo+0.05)
plt.ylim(score_prestazionale_minimo-0.025, score_prestazionale_massimo+0.025)

# Aggiungo titolo
plt.title('Scatterplot Score fisico vs prestazionale')

# Aggiungo legenda per il piede del giocatore
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[0], markersize=10, label='Mancino'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[1], markersize=10, label='Destro'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='black', markersize=10, label='Francesco Acerbi'),
                    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=inter_color, markersize=10, label='Altri giocatori dell\'Inter')],
           loc='lower right')

plt.show()
```
